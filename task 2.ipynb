```python
class FizzBuzz:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def play(self):
        for number in range(self.start, self.end + 1):
            if number % 3 == 0 and number % 5 == 0:
                print("Fizz Buzz")
            elif number % 3 == 0:
                print("Fizz")
            elif number % 5 == 0:
                print("Buzz")
            else:
                print(number)


# Creating Object of FizzBuzz Class
game = FizzBuzz(1, 100)
game.play()
```

```python
class MovieBudgetAnalyzer:
    def __init__(self, movies):
        self.movies = movies

    def add_movies(self):
        count = int(input("How many movies would you like to add? "))
        for _ in range(count):
            name = input("Enter movie name: ")
            budget = int(input("Enter movie budget: "))
            self.movies.append((name, budget))
        print("\nNew movies added successfully!\n")

    def calculate_average_budget(self):
        total_budget = sum(budget for _, budget in self.movies)
        return total_budget / len(self.movies)

    def analyze_budgets(self):
        average = self.calculate_average_budget()
        print(f"\nðŸŽ¬ Average Movie Budget: {average:,} USD\n")

        above_avg_count = 0
        print("ðŸ“ˆ Movies Above Average Budget:\n")
        for name, budget in self.movies:
            if budget > average:
                difference = budget - average
                print(f"â€¢ {name} â†’ {budget:,} USD  (ðŸ“Š +{difference:,} above average)")
                above_avg_count += 1

        print(f"\nðŸ’¡ Total movies above average budget: {above_avg_count}")

# ------- Main Program -------
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

app = MovieBudgetAnalyzer(movies)
app.analyze_budgets()
```
